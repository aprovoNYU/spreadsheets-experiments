# Download spreadsheets

## Download sample of spreadsheet files using requests and tqdm

```python
import requests
from requests.exceptions import HTTPError
import re
from datetime import datetime
import json
import csv
import os
from tqdm import tqdm
```

```python
with open('zenodo_spreadsheet_files_df_100recs_sample_2025-12-03_04-54_PM.csv', 'r') as f:
    sample_files = csv.DictReader(f)
    for rec in sample_files:
        print(rec)
        #https://stackoverflow.com/questions/48894060/how-to-print-the-value-of-a-specific-key-from-a-dictionary
        filekey = rec['filekey']
        print(filekey)
        filelink = rec['filelinks.self']
        print(filelink)
        rec_id = rec['id']
        print(rec_id)
        # for recid in rec:
            # print(recid)
        #https://www.geeksforgeeks.org/python/how-to-create-directory-if-it-does-not-exist-using-python/
        #https://www.slingacademy.com/article/python-requests-module-how-to-download-files-from-urls/#error-handling
        try:
            r = requests.get(filelink, stream=True)
            r.raise_for_status()
            if not os.path.exists(f'sample_spreadsheets_2025-12-03/{rec_id}'):
                os.makedirs(f'sample_spreadsheets_2025-12-03/{rec_id}')
            total = int(r.headers.get('content-length', 0))
            with open(f'sample_spreadsheets_2025-12-03/{rec_id}/{filekey}', 'wb') as file, tqdm(
                    desc='large_file.avi',
                    total=total,
                    unit='iB',
                    unit_scale=True,
                    unit_divisor=1024,
                ) as bar:
                for data in r.iter_content(chunk_size=1024):
                    size = file.write(data)
                    bar.update(size)
            print(f'Download of {filekey} complete.')
        except HTTPError as http_err:
            print(f'HTTP error occurred: {http_err}')
        except Exception as err:
            print(f'An error occurred: {err}')
print('Finished!')
```

## Download entire corpus using zenodo_get

```python
from zenodo_get import download
import csv
import re
import os
```

```python
#enter the filename
filepath = input("enter filepath")
```

```python
#https://www.geeksforgeeks.org/python-program-to-get-the-file-name-from-the-file-path/
filename = os.path.basename(filepath)
```

```python
print(filename)
```

```python
filetypes = ["*.csv","*.tsv","*.xlsx","*.xls","*.xlsb","*.xlsm","*.xltm","*.xltx","*.ods","*.ots","*.ogw","*.numbers"]
```

```python
with open(filename, 'r') as f:
    recs = csv.DictReader(f)
    for rec in recs:
        print(rec)
        recid=rec['id']
        print(recid)
        download(
            record_or_doi=recid,
            output_dir=f'./zenodo_get_test/{recid}',
            file_glob=filetypes,
        )
        #print(rec)
# with open file of ids:
# 	extension = get what’s before the first underscore
# 	extension = “*.”+extension
# 	recids = read the file as csv
# 	for each rec of recids:
# 		run zenodo_get with parameters row -o row -g *.extension
# 	save the terminal output of zenodo_get to file_report.txt
```
