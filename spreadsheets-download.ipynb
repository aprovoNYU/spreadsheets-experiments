# Download spreadsheets using zenodo_get

```python
import zenodo_get
import csv
import re
import os
```

```python
#enter the filename
filepath = input("enter filepath")
```

```python
#https://www.geeksforgeeks.org/python-program-to-get-the-file-name-from-the-file-path/
filename = os.path.basename(filepath)
```

```python
print(filename)
```

```python
filetypematch = re.match(r'[a-z]+', filename)
```

```python
filetype = filetypematch.group()
print(filetype)
```

```python
#listoffiles = ['tsv_small_sample_ids_2025-05-07_05-21_PM.csv','']
```

```python
#%%capture cap

with open(filename, 'r') as f:
    recids = csv.reader(f)
    for rec in recids:
        for recid in rec:
            print(recid)
            #zenodo_get.zenodo_get([recid, '-o' + recid, '-g*.' + filetype])
            zenodo_get.zenodo_get([recid, '-o' + recid, '-g*.xlsm', '-g*.csv'])
        #print(rec)
# with open file of ids:
# 	extension = get what’s before the first underscore
# 	extension = “*.”+extension
# 	recids = read the file as csv
# 	for each rec of recids:
# 		run zenodo_get with parameters row -o row -g *.extension
# 	save the terminal output of zenodo_get to file_report.txt
```
