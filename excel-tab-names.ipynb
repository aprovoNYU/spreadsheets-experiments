# Get Excel Tab Names for manual coding

```python
import pandas as pd
from datetime import datetime
```

```python
##https://stackoverflow.com/questions/17977540/pandas-looking-up-the-list-of-sheets-in-an-excel-file
def sheet_names(rec_id,filekey):
    try:
        xl = pd.ExcelFile(f'sample_spreadsheets_2025-12-03/{rec_id}/{filekey}')
        return xl.sheet_names
    except:
        return 'could not parse'
```

```python
sheet_names('14522238','Bufasi et al. (2024)_ Pre_post_results.xlsx')
```

```python
# import sample recs dataframe
#https://www.geeksforgeeks.org/python-pandas-series-from_csv/
sample_recs_df = pd.read_csv('zenodo_spreadsheet_files_df_100recs_sample_2025-12-03_04-54_PM.csv')
```

```python
# Source - https://stackoverflow.com/questions/17071871/how-do-i-select-rows-from-a-dataframe-based-on-column-values
# Posted by unutbu, modified by community. See post 'Timeline' for change history
# Retrieved 2025-12-03, License - CC BY-SA 4.0

sample_recs_df.loc[~sample_recs_df['filetype'].isin(['.csv','.tsv'])] # .loc is not in-place replacement
```

```python
#https://stackoverflow.com/questions/13331698/how-to-apply-a-function-to-two-columns-of-pandas-dataframe
# Source - https://stackoverflow.com/a
# Posted by ajrwhite, modified by community. See post 'Timeline' for change history
# Retrieved 2025-12-03, License - CC BY-SA 4.0

sample_recs_df['sheets'] = sample_recs_df.loc[~sample_recs_df['filetype'].isin(['.csv','.tsv'])].apply(lambda x: sheet_names(x.id, x.filekey), axis=1)
```

```python
sample_recs_df_exploded=sample_recs_df.explode('sheets')
```

```python
#https://www.slingacademy.com/article/pandas-how-to-concatenate-strings-from-multiple-columns-in-a-dataframe/
#https://stackoverflow.com/questions/44920796/concatenating-string-and-integers-in-pandas-dataframebased-on-conditions
sample_recs_df_exploded['sheetid'] = sample_recs_df_exploded['id'].astype(str) + '_' + sample_recs_df_exploded['filename_noext'] + '_' + sample_recs_df_exploded['sheets']
```

```python
sample_recs_df_exploded['id'] = sample_recs_df_exploded['id'].astype(str)
```

```python
sample_recs_df_exploded['sheetid'] = sample_recs_df_exploded[['id', 'filename_noext', 'sheets']].fillna('').agg('_'.join, axis=1).str.strip()
```
