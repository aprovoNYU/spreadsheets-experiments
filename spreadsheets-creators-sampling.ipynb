# Create a sampling frame for interviews from Zenodo spreadsheets corpus

```python
import re
from datetime import datetime
import json
import pandas as pd
import math
import numpy as np
from edtf import parse_edtf, struct_time_to_date
from decimal import Decimal
import matplotlib.pyplot as plt
```

```python
def parseable_date(x):
    y = parse_edtf(x)
    z = struct_time_to_date(y.lower_strict()).year
    return z
```

```python
#'/Volumes/ap180/zenodo_API_metadata_results_2025-01-08_01-46_PM.json'
#'zenodo_snippet.json'
with open('/Volumes/ap180/zenodo_API_metadata_results_2025-01-08_01-46_PM.json', 'r') as f:
    data = f.read()
zenodo_dict = json.loads(data)
```
